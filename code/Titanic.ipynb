```python
p = re.compile('Mrs|Mr|Miss|Master|Don|Dr|Ms|Major|Mlle|Mme|Countess|Lady')
```

```python
sir = []

for name in name_list :
	if p.search(name) != None :
		if p.search(name)[0] in ['Countess', 'Lady', 'Mrs'] :
			sir.append('Mrs')
		elif p.search(name)[0] in ['Mme', 'Ms', 'Mlle'] :
			sir.append('Miss')
		elif p.search(name)[0] in ['Don', 'Dr', 'Major'] :
			sir.append('Mr')
		else :
			sir.append(p.search(name)[0])
	else :
		sir.append('Other')
```

```python
# infant(0 ~ 5), child(6 ~ 12), teenager(13 ~ 18), youth(19 ~ 34),
# middle(35 ~ 49), seniors(50 ~ 64), elder(65 ~ )
```

```python
def category_age(age) :
	if age <= 0 and age < 6 :
		return 'infant'
	elif age <= 6 and age < 13 :
		return 'child'
	elif age <= 13 and age < 19 :
		return 'teenager'
	elif age <= 19 and age < 34 :
		return 'youth'
	elif age <= 35 and age < 50 :
		return 'middle'
	elif age <= 50 and age < 65 :
		return 'seniors'
	else :
		return 'elder'
```

```python
df_train['Age_cat'] = df_train['Age'].apply(category_age)
```

```python
df_train['Age_cat'].value_counts()
```

```python
pd.crosstab(df_train['Survived'], df_train['Age_cat'], margins=True)
```

```python
for index, value in enumerate(df_train['Age_cat'].unique()) :
	df_train['Age_cat'].replace(value, index, inplace=True)
```

```python
fare_mean = df_train['Fare'].mean()
```

```python
def category_fare(fare) :
	if fare > fare_mean :
		return 0
	else :
		return 1
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
```
